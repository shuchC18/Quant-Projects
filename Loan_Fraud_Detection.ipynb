# Encoding

Here, one-hot encoding is used. It converts categorical values into binary (0/1) columns so that machine learning models can process them.

**What does drop_first=True do?**

It drops the first category to avoid the dummy variable trapa situation where dummy variables are perfectly multicollinear, which can break linear models like Logistic Regression.

```python
x_enc = pd.get_dummies(df['purpose'], drop_first = True )

x_enc
```

# Bagging Classifier

 **What is BaggingClassifier?**

Bagging (Bootstrap Aggregating) is an ensemble learning method that builds multiple models (usually of the same type), each trained on a random subset of the training data (with replacement), and then aggregates their predictions (e.g., by majority vote for classification).

The BaggingClassifier in scikit-learn is a tool that helps you apply bagging to any base classifier (e.g., DecisionTreeClassifier).

 **Intuition Behind Bagging**

Why? A single model (like a decision tree) may overfit or capture noise.

How bagging helps: By training multiple models on different bootstrapped samples and combining them, the variance of the model is reduced.

 **How It Works**

Random sampling with replacement (bootstrapping) creates different datasets.

Train one model per dataset (often decision trees).

Aggregate predictions from all models:

Classification: Majority vote

Regression: Average

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
```

```python


bagging_class_log = BaggingClassifier(estimator=LogisticRegression(class_weight='balanced'),
                             n_estimators=20,max_samples=0.8,max_features=0.5,
                              bootstrap=True,bootstrap_features=False, random_state = 20)

bagging_class_log.fit(x_train, y_train)
```

```python
y_train_pred = bagging_class_log.predict(x_train)
y_test_pred = bagging_class_log.predict(x_test)
```

```python
print('Train confusion Matrix:')
sns.heatmap(confusion_matrix(y_train,y_train_pred),annot=True,fmt=".0f")
plt.show()
print('Test confusion Matrix:')
sns.heatmap(confusion_matrix(y_test,y_test_pred),annot=True,fmt=".0f")
plt.show()
```

```python
print('Train Classification Report:')
print(classification_report(y_train,y_train_pred))
print("-"*100)
print('Test Classification Matrix:')
print(classification_report(y_test,y_test_pred))
```

```python
scores = cross_val_score(bagging_class_log,x_train,y_train,scoring='recall',cv=5)
print('Score:',scores)
print('Avg Score:',np.mean(scores))
print('Std Score:',np.std(scores))
```
